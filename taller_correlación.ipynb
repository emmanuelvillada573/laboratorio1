```python
# Filtrar la variable objetivo 'abandono_cliente' para buscar correlaciones entre variables independientes
independent_corr = corr.drop(columns=['abandono_cliente']).drop(index=['abandono_cliente'])

# Encontrar pares de variables con correlación alta (valor absoluto > 0.9)
high_corr_pairs = []
for i in range(len(independent_corr.columns)):
    for j in range(i + 1, len(independent_corr.columns)):
        if abs(independent_corr.iloc[i, j]) > 0.9:
            pair = {
                'Variable 1': independent_corr.columns[i],
                'Variable 2': independent_corr.columns[j],
                'Correlación': independent_corr.iloc[i, j],
                '¿Son redundantes?': 'Sí' if abs(independent_corr.iloc[i, j]) > 0.9 else 'No'
            }
            high_corr_pairs.append(pair)

# Convertir a DataFrame para una mejor visualización
high_corr_df = pd.DataFrame(high_corr_pairs)

if not high_corr_df.empty:
    display(high_corr_df)
else:
    print("No se encontraron pares de variables independientes con una correlación (en valor absoluto) superior a 0.9.")
```

### PARTE 2 — Interpretación (Correlación de Pearson)

**Análisis conceptual**

**1. ¿Qué significa una correlación negativa en este contexto?**

En este contexto, una correlación negativa entre una variable independiente y `abandono_cliente` significa que a medida que el valor de la variable independiente aumenta, la probabilidad o tendencia de `abandono_cliente` disminuye. Por ejemplo, una correlación negativa fuerte entre `satisfaccion` y `abandono_cliente` indica que cuanto mayor es la satisfacción del cliente, menor es la probabilidad de que el cliente abandone.

**2. Explica la relación entre: satisfaccion y abandono_cliente**

La variable `satisfaccion` tiene una correlación de -0.875 con `abandono_cliente`. Esto es una correlación negativa muy fuerte. Significa que existe una relación inversa y casi lineal entre la satisfacción del cliente y el abandono. Específicamente, a medida que la satisfacción del cliente aumenta, la probabilidad de que un cliente abandone disminuye drásticamente. Esto sugiere que la satisfacción del cliente es un factor crucial para la retención y que mejorarla podría ser muy efectivo para reducir la tasa de abandono.

**3. Explica por qué una variable con correlación cercana a 0 no es útil.**

Una variable con una correlación cercana a 0 con la variable objetivo (en este caso, `abandono_cliente`) no es útil porque indica que no hay una relación lineal significativa entre ambas. Esto significa que los cambios en esa variable independiente no están linealmente asociados con los cambios en `abandono_cliente`. Para modelos lineales predictivos, esta variable aportaría muy poca o ninguna información para explicar la variabilidad de la variable objetivo, y por lo tanto, podría considerarse irrelevante para el modelo.

**4. ¿Por qué no debemos incluir variables altamente correlacionadas entre sí?**

No debemos incluir variables altamente correlacionadas entre sí (fenómeno conocido como **multicolinealidad**) en un modelo predictivo por varias razones:

*   **Redundancia:** Las variables altamente correlacionadas aportan información muy similar al modelo. Incluir ambas es redundante y no añade valor predictivo adicional significativo.
*   **Inestabilidad del modelo:** La multicolinealidad puede hacer que las estimaciones de los coeficientes de regresión sean inestables y difíciles de interpretar. Pequeños cambios en los datos pueden llevar a grandes cambios en los coeficientes.
*   **Dificultad de interpretación:** Es complicado determinar la contribución individual de cada variable a la predicción si están fuertemente relacionadas entre sí. Los errores estándar de los coeficientes tienden a ser altos, lo que hace que las variables parezcan no significativas incluso si en realidad lo son.
*   **Overfitting (sobreajuste):** En algunos casos, puede llevar a un modelo más complejo de lo necesario y propenso a sobreajustarse a los datos de entrenamiento.

Para evitar estos problemas, a menudo se selecciona solo una de las variables altamente correlacionadas o se utilizan técnicas de reducción de dimensionalidad.
